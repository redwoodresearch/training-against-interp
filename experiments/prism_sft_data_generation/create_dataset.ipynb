Load generated PRISM SFT data, filter to "base" model conversations, and push to HuggingFace.

```python
import json
import random
from pathlib import Path
from datasets import Dataset

NOTEBOOK_DIR = Path.home() / "training-against-interp" / "experiments" / "prism_sft_data_generation"

data = []
with open(NOTEBOOK_DIR / "results" / "prism_sft_dataset.jsonl") as f:
    for line in f:
        data.append(json.loads(line))

print(f"Loaded {len(data)} total records")
```

```python
# Filter to base model conversations only
base_records = [r for r in data if r["model"] == "base"]
print(f"Base model records: {len(base_records)}")

# Show an example
print(f"\nExample messages:")
for msg in base_records[0]["messages"]:
    print(f"  [{msg['role']}]: {msg['content'][:120]}...")
```

```python
# Build rows with just the messages field
random.seed(42)
rows = [{"messages": r["messages"]} for r in base_records]
random.shuffle(rows)

print(f"Total rows: {len(rows)}")
```

```python
# Create and inspect the dataset
ds = Dataset.from_list(rows)
print(ds)
print(ds[0])
```

```python
# Push to HuggingFace Hub
ds.push_to_hub("alignment-science/prism-base-sft-dataset")
```
